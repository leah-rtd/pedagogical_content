# Introduction au Machine Learning

Exemples concrets de tâches de Machine Learning
- Prédire le prix d'une maison
  - features = caractéristiques des maisons,
  - target = prix de vente
- Reconnaître un chiffre sur une image
  - features = chaque pixel de l'image
  - target = chiffre
- Trouver des thèmes récurrents dans un corpus de textes
  - features = textes
  - target = ? (il s'agit d'une tache non supervisée)

## Apprentissage supervisé  

- Les données sont étiquetées (on connaît les réponses)
- Objectif : prédire la target (variable cible)

## Apprentissage non supervisé  

- Pas de labels connus
- Objectif : Trouver des patterns cachés dans les données (ex : groupes ou motifs)

## Préparation de la donnée

- Transformer les données brutes pour qu'elles soint exploitables:
   - **One Hot Encoding** : transformer une feature catégorique en colonnes binaires -> les modèles ne savent pas comment intérpreter les chaines de charactères
   - **Mise à l'échelle** : mettre les features numériques à la meme échelle -> les modèles convergent mieux si les features sont mises à l'échelle et on ne veut pas que des features qui ont une plus grande échelle que d'autres influent plus sur le modèle

Exemple: Tache de classification binaire: chat ou chien ?  
Features **brutes**:  


|Couleur_poil|couleur_yeux|Hauteur(cm)|
|------------|------------|------------|
|marron|vert|30|
|blanc|vert|25|


Features **après** préparation:


|Couleur_poil_marron|Couleur_poil_blanc|couleur_yeux_vert|Hauteur(cm)|
|------------|------------|------------|------------|
|1|0|1|1|
|0|1|1|-1|

## Étapes  typiques d'un projet de Machine Learning Supervisé
1. Nettoyage des données
1. Sélection des features (quelles sont les variables qui sont vraiment importantes pour prédire ma target ?)
1. Préparation des données (encodage et mise à l'échelle)
1. Séparation en `X` (features) et `y` (target)
2. Séparation de `X_train` et `y_train` (jeu d'entrainement) et `X_test` et `y_test` (jeu de test)
1. Choix du modèle
1. Entraînement sur le jeu d'entrainement
1. Évaluation sur le jeu de test

### Pourquoi évaluer les modèles ?

Le problème du **surapprentissage**
- Un modèle peut apprendre par cœur les patterns des données d'entraînement
- Dans ce cas, il ne peut pas généraliser sur des données non vues
- **Solution** : Tester sur des données non vues pendant l'entraînement pour comprendre s'il peut généraliser
